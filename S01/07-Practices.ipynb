Here are **20 fun and beginner-friendly Python problems** with **short, comment-free solutions**, based on what you've taught so far (`print`, `input`, `variables`, `random`). They are simple enough for students learning early-stage Python.

### 🧠 Problems and Answers

#### 1. Greet the user by name

```python
name = input()
print("Hello", name)
```

#### 2. Add two numbers

```python
a = int(input())
b = int(input())
print(a + b)
```

#### 3. Print a random number between 1 and 10

```python
import random
print(random.randint(1, 10))
```

#### 4. Print double the user's number

```python
n = int(input())
print(n * 2)
```

#### 5. Choose a random color

```python
import random
colors = ['red', 'blue', 'green', 'yellow']
print(random.choice(colors))
```

#### 6. Repeat user’s name 3 times

```python
name = input()
print(name * 3)
```

#### 7. Create a silly username

```python
name = input()
pet = input()
print(name + pet + "123")
```

#### 8. Print a number plus a random number

```python
import random
n = int(input())
print(n + random.randint(1, 10))
```

#### 9. Simple dice roll

```python
import random
print("🎲 You rolled:", random.randint(1, 6))
```

#### 10. Calculate age in months

```python
age = int(input())
print(age * 12)
```

#### 11. Print a fun pattern

```python
print("😊" * 5)
```

#### 12. Create a fake hacker name

```python
name = input()
print("xX_" + name + "_Xx")
```

#### 13. Add and multiply two numbers

```python
a = int(input())
b = int(input())
print((a + b) * 2)
```

#### 14. Random fruit picker

```python
import random
fruits = ['apple', 'banana', 'cherry', 'kiwi']
print(random.choice(fruits))
```

#### 15. Swap two variables

```python
a = input()
b = input()
a, b = b, a
print(a, b)
```

#### 16. Guess the number game (one guess)

```python
import random
n = random.randint(1, 5)
g = int(input())
print("You win!" if g == n else "Wrong!")
```

#### 17. Name length counter

```python
name = input()
print(len(name))
```

#### 18. Mix two words

```python
a = input()
b = input()
print(a[:2] + b[-2:])
```

#### 19. Convert Celsius to Fahrenheit

```python
c = float(input())
print(c * 9/5 + 32)
```

#### 20. Make a secret code by reversing name

```python
name = input()
print(name[::-1])
```



### 🧠 20 Fun & Cool Python Problems for Beginners

#### 🧮 Basics & Input

1. **Name Mirror**

   * Ask for the user's name and print it backward.
   * *Example: “Alice” → “ecilA”*

2. **Silly Sentence Generator**

   * Ask for a noun, verb, and adjective and insert them into a funny sentence.
   * *"The {adjective} {noun} likes to {verb} all day."*

3. **Magic 8 Ball**

   * User asks a yes/no question.
   * Use `random.choice()` to give random answers like: "Yes", "No", "Maybe", "Ask again later".

4. **Age in Dog Years**

   * Ask the user for their age and convert it to dog years (×7).

5. **Username Generator**

   * Input first name and favorite number, then generate a "cool" username.
   * *Example: “Lily42”*

---

#### 🎲 Random Fun

6. **Dice Roller**

   * Simulate rolling two 6-sided dice using `random.randint(1, 6)` and print the result.

7. **Lucky Number**

   * Ask for user's name, and generate a “lucky number” using `random.randint(1, 100)`.

8. **Random Password Generator**

   * Generate a 6-character password using random letters/numbers from a list.

9. **Rock, Paper, Scissors (Single Round)**

   * User picks one; the computer picks randomly.
   * Print who wins.

10. **Magic Fortune Cookie**

* When run, print a random fortune from a list of 10 fortunes.

---

#### 🧠 Math & Numbers

11. **Simple Calculator**

* Ask for two numbers and print their sum, difference, product, and division.

12. **Tip Calculator**

* Ask for bill amount and tip %, then calculate and print the total.

13. **Guess My Number (One Attempt)**

* Generate a number 1–10. User guesses it. Print "Correct!" or "Wrong."

14. **Favorite Number Game**

* Ask for the user's favorite number and tell them if it’s even or odd.

15. **Birthday Countdown**

* Ask for age and print how many years until 100.

---

#### 💬 String & Fun Outputs

16. **Funny Echo**

* Ask for a word and print it 3 times with different effects.
* *“Hello” → hello, HeLlO, HELLO!*

17. **Emoji Mood**

* Ask how they feel today and match a response with an emoji.
* *Sad → 😢, Happy → 😊, Tired → 😴*

18. **Pig Latin Translator**

* Ask for a word and convert it to Pig Latin.
* *“apple” → “appleyay”, “dog” → “ogday”* (optional challenge)

19. **ASCII Banner**

* Ask for their name and print it like:

  ```
  * L *
  * i *
  * l *
  * y *
  ```

20. **Countdown Printer**

* Ask for a number and print a countdown to zero using `print()`.
* *No loops! Use chained print statements or `input()` for each step.*
