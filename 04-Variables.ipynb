# Python Variables

A **variable** is a name that stores a value. You can use variables to remember information and use it later in your code.

---

## Creating Variables

To create a variable, just choose a name and use the `=` sign to assign a value.

```python
age = 10
name = "Alice"
height = 1.65
```

---

## Using Variables

You can use variables in calculations or print them out.

```python
age = 10
print(age + 5)      # Output: 15

name = "Alice"
print("Hello, " + name)  # Output: Hello, Alice
```

---

## Rules for Variable Names

- Use letters, numbers, and underscores (`_`)
- Cannot start with a number
- Names are case-sensitive (`Age` and `age` are different)

**Examples:**

```python
my_number = 7
user_name = "Bob"
```

---

## Changing Variable Values

You can change the value of a variable by assigning it a new value.

```python
score = 100
score = 120
print(score)   # Output: 120
```
